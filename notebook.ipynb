```python
for col in df.select_dtypes(include=['int64', 'float64']).columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower=lower, upper=upper)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Input
tmax = int(input("Enter the maximum time: "))
initial = float(input("Enter the initial water level (m): "))
rise = float(input("Enter the rise in water level per unit time (m): "))

# Calculation
t = np.arange(0, tmax + 1)
water = initial + rise * t

# Output
print("Time values:")
print(t)

print("\nWater levels:")
print(water)

# Plot
plt.figure(figsize=(8,5))
plt.plot(t, water, 'bo-', linewidth=2)
plt.grid(True)
plt.xlabel("Time")
plt.ylabel("Water Level (m)")
plt.title("Rising Water Level")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Input
tmax = int(input("Enter the maximum time: "))
initial = float(input("Enter the initial water level (m): "))
rise = float(input("Enter the rise in water level (m): "))

# Calculation
t = np.arange(0, tmax + 1)
water = initial + rise * t

# Output
print("Time values:")
print(t)

print("\nWater levels:")
print(water)

# Bar Graph
plt.figure(figsize=(8, 5))
plt.bar(t, water)
plt.grid(True)
plt.xlabel("Time")
plt.ylabel("Water Level (m)")
plt.title("Rising Water Level - Bar Graph")
plt.show()
```
